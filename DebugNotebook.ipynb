```python
import pandas as pd
  
# read specific columns of csv file using Pandas
df = pd.read_csv("/home/omarelseadawy/Desktop/FinalThesisIsa/graphing/ResultsCombines.csv")
print(df)
```

```python
APXVals = df.APXNON.values.tolist()
APXVals = APXVals[:200]
```

```python
PPOVals = df.PPO.values.tolist()
PPOVals = PPOVals[:200]
```

```python
DQNVals = df.DQN.values.tolist()
DQNVals = DQNVals[:200]
```

```python
## OLD WAS MARWIL, now this is preemptive version
MARVals = df.APXPRE.values.tolist()
MARVals = MARVals[:200]
```

```python
ARSVals = df.ARS.values.tolist()
ARSVals = ARSVals[:200]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(18, 10))
try:
    plt.plot(range(0,len(APXVals)),APXVals,'-b',markersize=1,label='Ape-XNon')
    plt.plot(range(0,len(PPOVals)),PPOVals,'--r',markersize=1,label='PPO')
    plt.plot(range(0,len(DQNVals)),DQNVals,'-.go',markersize=1,label='DQN')
    plt.plot(range(0,len(ARSVals)),ARSVals,':mo',markersize=1,label='ARS')
    plt.plot(range(0,len(MARVals)),MARVals,'-co',markersize=1,label='Ape-XPre')
#     plt.yscale("log")
except ValueError:
        pass
plt.ylabel('Mean Reward')
plt.xlabel('Iterations')
plt.ylim([-200,300])
plt.title("Training Phase")
plt.legend(loc="lower right")
plt.savefig("Training.png")
plt.show()
```
